```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('SAR_raw_data_2.csv')
```

```python
data = data.drop('Unnamed: 0', axis = 1)
```

```python
is_nan_data = (data.shape[0]-data.isna().sum()).sort_values()
```

```python
start_data = data[is_nan_data[is_nan_data > 400].index]
```

```python
start_data.describe()
```

```python
drop_index = start_data[start_data['htherm_sar'].isna() == True].index
start_data = start_data.drop(axis=0, index = drop_index)
```

Обработка данных, восстановление пропущенных данных

```python
start_data.isna().sum()
```

```python
for column in start_data.columns:
    print(start_data[column].value_counts().nlargest(1))
```

```python
start_data['squid_temp'] = start_data['squid_temp'].fillna(300.0)
```

```python
start_data[['squid_rem_mag','squid_coerc_f']] = start_data[['squid_rem_mag','squid_coerc_f']].fillna(0)
```

```python
start_data[['emic_z_size','emic_x_size','emic_y_size']] = start_data[['emic_z_size','emic_x_size','emic_y_size']].fillna(start_data[['emic_z_size','emic_x_size','emic_y_size']].mean())
```

```python
start_data['emic_shape'] = start_data['emic_shape'].fillna('NaN')
```

```python
start_data['emic_size'] = start_data['emic_size'].fillna(start_data['emic_size'].mean())
```

```python
start_data = start_data.drop_duplicates()
```

```python
start_data = start_data.drop(axis = 0, index =start_data[start_data['paper_doi'].isna()==True].index)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import GridSearchCV
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python

model = RandomForestRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python

print(mean_absolute_error(y_pred, y_test))
print(mean_squared_error(y_pred, y_test))
```

```python
param_grid = {'n_estimators' : [100,300,500,1000],
             'min_samples_split': [2,4,6,7],
             'min_samples_leaf': [2, 4, 6, 10]}
gs = GridSearchCV(estimator=model, param_grid = param_grid, verbose =1, cv = 5)
gs.fit(X_train, y_train)
print(gs.best_estimator_)
y_pred = gs.predict(X_test)
print(mean_absolute_error(y_pred, y_test))
print(mean_squared_error(y_pred, y_test))
```
